```python
import pandas as pd

# Updated with the correct file name found from the !ls command
data_path = '/content/Online_data/OnlineOrders_of_a_ecommerce_website.csv'

try:
    df = pd.read_csv(data_path)
    print("Data types:")
    print(df.dtypes)

    print("\nShape of the DataFrame:")
    print(df.shape)

    print("\nFirst 5 rows of the DataFrame:")
    display(df.head())
except FileNotFoundError:
    print(f"Error: The file '{data_path}' was not found.")
```

```python
print("### Data Information ###")
display(df.info())

print("\n### Missing Values Check ###")
missing_values = df.isnull().sum()
display(missing_values[missing_values > 0])

print("\n### Duplicate Rows Check ###")
duplicates_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates_count}")

print("\n### Unique Values Count ###")
display(df.nunique())

# Fixed the SyntaxError by adding the closing parenthesis
df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'], errors='coerce')
df['crawl_timestamp'].dtype
df[df['crawl_timestamp'].isnull()]
```

```python
# Extracting date and year from the timestamp
df['crawl_date'] = df['crawl_timestamp'].dt.date
df['crawl_year'] = df['crawl_timestamp'].dt.year

# Displaying the first 5 rows to see the new columns
print("New columns 'crawl_date' and 'crawl_year' added:")
display(df[['crawl_timestamp', 'crawl_date', 'crawl_year']].head())
df["month"] = df["crawl_timestamp"].dt.month
display(df[['month']].head())
```

```python
# Applying strip to remove leading/trailing spaces
df['product_name'] = df['product_name'].str.strip()

print("Product names have been stripped of extra spaces.")
print("\nSample of cleaned product names:")
display(df['product_name'].head())
df['product_name'] = df['product_name'].str.lower()
df['product_name'].value_counts().head()
```

```python
# Counting occurrences of each main category
category_counts = df['main_category'].value_counts()

print("Number of products per main category:")
display(category_counts)
```

```python
df['product_category_tree'] = (
    df['product_category_tree']
    .str.replace('[','', regex=False)
    .str.replace(']','', regex=False)
    .str.replace('"','', regex=False)
)

print("Cleaned product_category_tree sample:")
display(df['product_category_tree'].head())
```

```python
df['main_category'] = df['product_category_tree'].str.split(' >> ').str[0]

print("Main category extracted successfully.")
print("\nFirst 5 entries of main_category:")
display(df['main_category'].head())
```

```python
# Counting occurrences of each main category
category_counts = df['main_category'].value_counts()

# Identifying categories with only one entry
small_categories = category_counts[category_counts == 1]
print(f"Number of categories with only one product: {len(small_categories)}")

if len(small_categories) > 0:
    print("\nSample of these small categories:")
    display(small_categories.head(10))
```

```python
valid_categories = df['main_category'].value_counts()

df['main_category'] = df['main_category'].apply(
    lambda x: x if valid_categories[x] > 10 else 'Other'
)

print("Categories with 10 or fewer items have been grouped into 'Other'.")
print("\nUpdated value counts for main_category:")
display(df['main_category'].value_counts())
```

```python
def detect_outliers(group):
    Q1 = group['retail_price'].quantile(0.25)
    Q3 = group['retail_price'].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return group[(group['retail_price'] < lower) | (group['retail_price'] > upper)]

outliers = df.groupby('main_category', group_keys=False).apply(detect_outliers)

print("Total outliers:", len(outliers))
print("\nOutlier count per category:")
display(outliers['main_category'].value_counts())

print("\nSample of detected outliers:")
display(outliers[['product_name', 'main_category', 'retail_price']].head())
```

```python
# Checking for duplicates based on product name and brand
duplicate_subset_count = df.duplicated(subset=['product_name', 'brand']).sum()

print(f"Number of duplicate rows based on product_name and brand: {duplicate_subset_count}")

if duplicate_subset_count > 0:
    print("\nSample of these duplicates:")
    display(df[df.duplicated(subset=['product_name', 'brand'], keep=False)].sort_values(by='product_name').head(10))
```

```python
# Calculate discount percentage
df['discount_percent'] = ((df['retail_price'] - df['discounted_price']) / df['retail_price']) * 100

print("### Discount Percentage Statistics ###")
display(df['discount_percent'].describe())

# Identify negative discounts
negative_discounts = df[df['discount_percent'] < 0]
print(f"\nNumber of products with negative discount: {len(negative_discounts)}")
if len(negative_discounts) > 0:
    display(negative_discounts[['product_name', 'retail_price', 'discounted_price', 'discount_percent']].head())

# Identify extremely high discounts (> 90%)
high_discounts = df[df['discount_percent'] > 90]
print(f"\nNumber of products with > 90% discount: {len(high_discounts)}")
if len(high_discounts) > 0:
    display(high_discounts[['product_name', 'retail_price', 'discounted_price', 'discount_percent']].head())
```

```python
df['sub_category'] = df['product_category_tree'].str.split(' >> ').str[1]

print("Sub-category extracted successfully.")
print("\nFirst 5 entries of sub_category:")
display(df[['product_category_tree', 'sub_category']].head())
```

```python
df['price_outlier'] = df.index.isin(outliers.index)

print("Price outliers have been flagged in the main DataFrame.")
print(f"Total outliers flagged: {df['price_outlier'].sum()}")

print("\nSample of rows marked as outliers:")
display(df[df['price_outlier'] == True].head())
```

```python
import re

def clean_product_name(name):
    if not isinstance(name, str):
        return name
    name = name.lower()                # lowercase
    name = re.sub(r'[^\w\s]', '', name) # remove punctuation
    name = re.sub(r'\d+', '', name)     # remove numbers
    name = re.sub(r'\s+', ' ', name)    # remove extra spaces
    return name.strip()

df['clean_product_name'] = df['product_name'].apply(clean_product_name)

print("Product names have been deeply cleaned (lowercase, no punctuation, no numbers).")
print("\nComparison of original vs cleaned:")
display(df[['product_name', 'clean_product_name']].head())
```

```python
stop_words = ['for','with','and','the','new','pack','set']

def remove_stopwords(text):
    if not isinstance(text, str):
        return text
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

df['clean_product_name'] = df['clean_product_name'].apply(remove_stopwords)

print("Stop words have been removed from the product names.")
print("\nSample of final cleaned names:")
display(df[['product_name', 'clean_product_name']].head())
```

```python
from rapidfuzz import process, fuzz

# Getting unique cleaned product names
names = df['clean_product_name'].unique()

duplicates = []

# We'll limit this to a sample if the dataset is too large,
# but here we'll try the logic on the unique names found.
print(f"Processing {len(names)} unique names for fuzzy matching...")

# Using a small slice for demonstration if needed, or run on full set
for name in names[:500]:  # Limiting to first 500 unique names for performance in this step
    matches = process.extract(name, names, scorer=fuzz.ratio, limit=5)

    for match in matches:
        # match is (string, score, index)
        if match[1] > 90 and match[0] != name:
            duplicates.append((name, match[0], match[1]))

# Displaying the first 10 fuzzy duplicates found
import pandas as pd
fuzzy_df = pd.DataFrame(duplicates, columns=['Name A', 'Name B', 'Score'])
print(f"\nFound {len(fuzzy_df)} potential fuzzy duplicates (sample size: 500).")
display(fuzzy_df.head(10))
```

## Data Cleaning and Fixing

### Subtask:
Clean the dataset by removing duplicates, handling price outliers, and ensuring correct data types.


**Reasoning**:
I will clean the dataset by removing duplicates based on 'product_id', removing the flagged price outliers, ensuring numeric columns are floats, and dropping intermediate helper columns to finalize the cleaning phase.



```python
import pandas as pd

# 1. Remove duplicate rows based on 'product_id', keeping the first occurrence
df_cleaned = df.drop_duplicates(subset=['product_id'], keep='first').copy()

# 2. Address price outliers using the 'price_outlier' flag (removing them as requested)
df_cleaned = df_cleaned[df_cleaned['price_outlier'] == False]

# 3. Verify and cast numeric columns to float types
df_cleaned['retail_price'] = df_cleaned['retail_price'].astype(float)
df_cleaned['discounted_price'] = df_cleaned['discounted_price'].astype(float)

# 4. Drop unnecessary intermediate columns
# Keeping main columns, removing temporary flags or cleaning buffers
columns_to_drop = ['price_outlier', 'clean_product_name']
df_cleaned = df_cleaned.drop(columns=columns_to_drop)

# 5. Print the shape of the cleaned DataFrame
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_cleaned.shape}")
print("\nFirst 5 rows of cleaned data:")
display(df_cleaned.head())

# Updating the reference for the next steps
df = df_cleaned
```
