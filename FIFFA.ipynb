```python
import pandas as pd

# Path to the extracted csv file
csv_path = '/content/Fiffa_extracted/fifa21 raw data v2.csv'

# Load the data
df = pd.read_csv(csv_path)

# Show requested info
print("--- Head ---")
display(df.head())

print("\n--- Info ---")
df.info()

print("\n--- Shape ---")
print(df.shape)
```

```python
# Drop the 'positions' column
df = df.drop(columns=['positions'], errors='ignore')

print('--- Columns after dropping positions ---')
print(f"'positions' in df: {'positions' in df.columns}")
```

```python
# Final check for any remaining missing values across all columns
print('--- Final Sorted Missing Values Count ---')
missing_stats = df.isnull().sum().sort_values(ascending=False)
display(missing_stats[missing_stats > 0] if (missing_stats > 0).any() else 'No missing values found!')
```

```python
# Convert id to categorical (string)
df['id'] = df['id'].astype(str)

# Identify numerical and categorical columns
num_cols = df.select_dtypes(include=['number']).columns.tolist()
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f'Numerical Columns ({len(num_cols)}):', num_cols)
print(f'Categorical Columns ({len(cat_cols)}):', cat_cols)

print('\n--- Numerical Columns Summary ---')
display(df.describe())

print('\n--- Categorical Columns Summary ---')
display(df.describe(include=['object']))
```

```python
# Fill missing values in contract_end with 'Unknown'
df['contract_end'] = df['contract_end'].fillna('Unknown')

print('--- Check Missing Values in contract_end ---')
print(f"Missing values in 'contract_end': {df['contract_end'].isnull().sum()}")
display(df[['name', 'contract', 'contract_end']].head())
```

```python
print("--- Columns ---")
print(df.columns.tolist())

print("\n--- Missing Values (Sorted) ---")
missing_values = df.isnull().sum().sort_values(ascending=False)
display(missing_values[missing_values > 0])
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("/", "_")
    .str.replace("-", "_")
    .str.replace("↓", "")
)

print("--- Updated Column Names ---")
print(df.columns.tolist())
```

```python
df = df.rename(columns={
    "ova": "overall_rating",
    "pot": "potential",
    "w_f": "weak_foot",
    "a_w": "attacking_work_rate",
    "d_w": "defensive_work_rate",
    "longname": "long_name"
})

print("--- Renamed Columns ---")
print(df.columns.tolist())
```

```python
# Drop photourl and playerurl columns safely
columns_to_drop = ['photourl', 'playerurl']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns], errors='ignore')

print('--- Columns after dropping URLs ---')
print(f"'photourl' in df: {'photourl' in df.columns}")
print(f"'playerurl' in df: {'playerurl' in df.columns}")
```

```python
num_cols = len(df.columns)
mid_point = num_cols // 2

print('--- First 10 Columns ---')
display(df.iloc[:, :10].head())

print('\n--- Middle 10 Columns ---')
display(df.iloc[:, mid_point-5 : mid_point+5].head())

print('\n--- Last 10 Columns ---')
display(df.iloc[:, -10:].head())
```

```python
# Clean the 'club' column by removing newlines and whitespace
df['club'] = df['club'].str.replace('\n', '', regex=True).str.strip()

print('--- Unique Club Names (First 10) ---')
print(df['club'].unique()[:10])

display(df[['name', 'club']].head())
```

```python
# Split the contract column into start and end years
df[['contract_start', 'contract_end']] = df['contract'].str.split('~', expand=True)

# Strip whitespace from the new columns
df['contract_start'] = df['contract_start'].str.strip()
df['contract_end'] = df['contract_end'].str.strip()

print('--- Contract Columns Sample ---')
display(df[['contract', 'contract_start', 'contract_end']].head())
```

```python
# Display a sample of the positions column
print('--- Positions Column Sample ---')
display(df['positions'].head(10))
```

```python
# Fix: Only split if the column contains strings to avoid NaN results
import pandas as pd

if isinstance(df['positions'].iloc[0], str):
    df['positions'] = df['positions'].str.split(',')
    print('--- Positions Column Successfully Split ---')
else:
    print('--- Positions Column Already Processed as List or Contains Non-String Data ---')

display(df['positions'].head(10))
```

```python
def convert_value(x):
    if isinstance(x, (int, float)):
        return x
    x = x.replace('€','')

    if 'M' in x:
        return float(x.replace('M','')) * 1_000_000
    elif 'K' in x:
        return float(x.replace('K','')) * 1_000
    else:
        return float(x)

df['value'] = df['value'].apply(convert_value)

print('--- Value Column Sample (Numeric) ---')
display(df[['name', 'value']].head())
df['wage'] =df['wage'].str.replace("€", "").str.replace("K", "000").astype(float)
```

```python
df['release_clause'] = df['release_clause'].apply(convert_value)

print('--- Release Clause Column Sample (Numeric) ---')
display(df[['name', 'release_clause']].head())
```

```python
def height_to_cm(x):
    if "'" in str(x):
        feet, inches = x.split("'")
        inches = inches.replace('"','')
        return int(feet)*30.48 + int(inches)*2.54
    return x

df['height_cm'] = df['height'].apply(height_to_cm)

print('--- Height Conversion Sample ---')
display(df[['name', 'height', 'height_cm']].head())
```

```python
def convert_weight(x):
    if 'lbs' in str(x):
        return float(x.replace('lbs', '')) * 0.453592
    elif 'kg' in str(x):
        return float(x.replace('kg', ''))
    return x

df['weight_kg'] = df['weight'].apply(convert_weight)

print('--- Weight Conversion Sample (Fixed) ---')
display(df[['name', 'weight', 'weight_kg']].head())
```

```python
# Check missing values in hits
print(f"Missing values in 'hits': {df['hits'].isnull().sum()}")

def convert_hits(x):
    if isinstance(x, str) and 'K' in x:
        return float(x.replace('K','')) * 1000
    try:
        return float(x)
    except (ValueError, TypeError):
        return x

df['hits'] = df['hits'].apply(convert_hits)

print('\n--- Hits Column Sample (Numeric) ---')
display(df[['name', 'hits']].head())
```

```python
# Fill missing values in loan_date_end and hits
df['loan_date_end'] = df['loan_date_end'].fillna('No Loan')
df['hits'] = df['hits'].fillna(0)

# Final check for missing values
print('--- Final Missing Values Count ---')
print(df.isnull().sum().sort_values(ascending=False).head(10))
df[df['positions'].isnull()][['name','positions']]
```
